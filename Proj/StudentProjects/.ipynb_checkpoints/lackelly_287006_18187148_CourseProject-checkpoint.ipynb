## Introduction

The Olivetti Faces dataset was created in the span of two years from April 1992 - April 1994 by the AT&T Laboratories of Cambridge. The dataset consists of ten different images of 40 distinct people each, with each picture taken under different circumstances (i.e. lighting, facial expressions, facial features). In this project, I want to analyze the Olivetti faces dataset using Principal Component Analysis (PCA) to simplify the machine learning process and hopefully, find some trends within these images and data.

## Main portion of the project

```
# Importing the libraries necessary for this project

import numpy as np
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```
%matplotlib inline
```

```
olivetti_faces = fetch_olivetti_faces() # Download dataset
data = olivetti_faces.data.reshape(400,64,64) # Data had to be reshapen into a 3D array
target = olivetti_faces.target
```

```
data.shape
```

```
target.shape
```

Now that I have established the basic variables of the dataset, I'm going to display the 40 distinct faces, as well as a sample of the first 4 people's various pictures.

```
def distinct_faces(images, people):
    fig, axs = plt.subplots(nrows=4, ncols=10, figsize=(18,9))
    axs = axs.flatten()

    for person in range(40):
        index = person*10
        axs[person].imshow(images[index], cmap="gray")
        axs[person].set_xticks([])
        axs[person].set_yticks([])
        axs[person].set_title("Person {}".format(person))
    plt.suptitle("There are 40 distinct people in this dataset")
```

```
distinct_faces(data,target)
```

```
def distinct_faces_sample(images, people):
    fig, axs = plt.subplots(nrows=4, ncols=10, figsize=(18,9))

    i=0
    for person in people:
        for j in range(10):
            index = person*10 + j
            axs[i,j].imshow(images[index], cmap="gray")
            axs[i,j].set_xticks([])
            axs[i,j].set_yticks([])
            axs[i,j].set_title("Person {}".format(person))
        i = i+1
    plt.suptitle("The first 4 people's various pictures")
```

```
distinct_faces_sample(data,[0,1,2,3]) # Can change [] for different people
```

#### Using PCA in Machine Learning

I'm going to first import the dataset into a Pandas DataFrame, then standardize the data with StandardScaler, as PCA is better performed under standardized data.

```
df = pd.DataFrame(data = olivetti_faces.data.transpose(), columns = olivetti_faces.target)
```

```
df.head()
```

```
x = df.loc[:,].values
```

```
y = df.loc[:,0:39.9]
```

```
x = StandardScaler().fit_transform(x)
```

```
pca = PCA (n_components = 2) # project to 2D
pComponents = pca.fit_transform(x)
pDf = pd.DataFrame(data = pComponents, columns = ["Principcal Component 1","Principal Component 2"])
```

```
pDf.head()
```

```
alt.Chart(pDf).mark_circle().encode(
    x = df.loc[:,].values,
    y = df.loc[:,1:2]
)
```
